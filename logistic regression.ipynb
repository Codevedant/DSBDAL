```python
x = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,
random_state = 0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print("Transformed x_train Values:\n\n", x_train)
print("\n\nTransformed x_train Values:\n\n", x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
y_pred = log_reg.predict(x_test)
print("Predicted Values:\n\n", y_pred)
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report
)
```

```python
class_report = classification_report(y_test, y_pred)
```

```python
print("\nClassification Report:\n\n", class_report)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n\n", conf_matrix)
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
precision = precision_score(y_test, y_pred)
print("Precision:", precision)
```

```python
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
```

```python
error_rate = 1 - accuracy
print("Error rate:", error_rate)
```
